```python
#import data for telangana rera

# Imports
import pandas as pd
from sqlalchemy import create_engine

# This CSV doesn't have a header so pass
# column names as an argument
columns = [
'project_id', 'project_name', 'project_status',
       'proposed_completion_date', 'project_type', 'total_building_unit',
       'total_area', 'street', 'locality', 'state', 'division', 'district',
       'taluka', 'village', 'pincode', 'approved_build_up_area'
    
]

# Load in the data
df = pd.read_csv(
    "/home/anm/rera_maha/final_maha/pune/pune_final.csv"  ,names=columns  
)

# Instantiate sqlachemy.create_engine object
engine = create_engine('postgresql://postgres:2020@localhost:5432/Real Estate')

# Save the data from dataframe to
# postgres table "iris_dataset"
df.to_sql(
    'projects', 
    engine,
    index=False, # Not copying over the index
    if_exists='append'
   
)
```
